```python
import pandas as pd
from openpyxl import load_workbook
from pathlib import Path

src = Path(r"C:\Code_Learning\repos\Soccer_Players_Market_Value\Data\Corrections_Log.xlsx")
dst = src.with_name("Corrections_Log_recovered.xlsx")

# 1) Read all sheets with pandas
xls = pd.ExcelFile(src)
sheet_names = xls.sheet_names
print("Sheets:", sheet_names)

# 2) Load with openpyxl to inspect workbook structure
wb = load_workbook(src)
print("Openpyxl sheets:", wb.sheetnames)

# 3) Copy sheet contents into a new workbook
with pd.ExcelWriter(dst, engine="openpyxl") as writer:
    for sh in sheet_names:
        df = pd.read_excel(src, sheet_name=sh)
        df.to_excel(writer, sheet_name=sh[:31], index=False)

print(f"Saved recovered copy to: {dst}")
```
